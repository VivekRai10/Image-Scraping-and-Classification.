```python
#connecting to the webdriver
driver = webdriver.Chrome(r"C:\Users\saurabh\Downloads\chromedriver_win32 (1)\chromedriver.exe")
```

### scraping images of sarees

```python
#specifying the url of the webpage to be scraped
url="https://www.amazon.com/s?k=sarees&ref=nb_sb_noss_1"
```

```python
driver.get(url)
```

```python
#getting the list of url of sarees
sarees=[]

for j in range(0,6):

    sari=driver.find_elements_by_xpath("//div[@class='a-section aok-relative s-image-square-aspect']")
    for i in sari:
        dataset1=i.find_element_by_tag_name("img")
        saree=sarees.append(dataset1.get_attribute('src'))
    button=driver.find_element_by_xpath("//*[contains(text(), 'Next')]")
    driver.execute_script("arguments[0].click();", button)
    time.sleep(5)
```

```python
sarees
```

```python
import requests
for img in sarees:
 file_name = img.split('/')[-1]
 print("Downloading file:%s"%file_name)
 r = requests.get(img, stream=True)
 with open(file_name, 'wb') as f:
    for chunk in r:
       f.write(chunk)
```
